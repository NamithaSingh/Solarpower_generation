```python
# MODEL BUILDING
# 1. RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

x = df.drop(columns=['powergenerated'])
y = df['powergenerated']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
x_train, x_test, y_train, y_test
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred
```

```python
mae_rf = mean_absolute_error(y_test, y_pred)
mse_rf = mean_squared_error(y_test, y_pred)
r2_rf = r2_score(y_test, y_pred)
```

```python
print(f"Mean Absolute Error (MAE): {mae_rf}")
print(f"Root Mean Squared Error (RMSE): {mse_rf}")
print(f"R-squared (R2): {r2_rf}")
```

```python
# 2. XGBoost (Extreme Gradient Boosting)

from xgboost import XGBRegressor

xgb_model = XGBRegressor(n_estimators=100, random_state=42)
xgb_model.fit(x_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(x_test)
y_pred_xgb
```

```python
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)
```

```python
print(f"Mean Absolute Error (MAE): {mae_xgb}")
print(f"Root Mean Squared Error (RMSE): {mse_xgb}")
print(f"R-squared (R2): {r2_xgb}")
```

```python
# 3. LightGBM (Light Gradient Boosting Machine)

from lightgbm import LGBMRegressor

lgbm_model = LGBMRegressor(n_estimators=100, random_state=42)
lgbm_model.fit(x_train, y_train)
```

```python
y_pred_lgbm = lgbm_model.predict(x_test)
y_pred_lgbm
```

```python
mae_lgbm = mean_absolute_error(y_test, y_pred_lgbm)
mse_lgbm = mean_squared_error(y_test, y_pred_lgbm)
r2_lgbm = r2_score(y_test, y_pred_lgbm)
```

```python
print(f"Mean Absolute Error (MAE): {mae_lgbm}")
print(f"Root Mean Squared Error (RMSE): {mse_lgbm} ")
print(f"R-squared (R2): {r2_lgbm}")
```

```python
# 4. SVR (Support Vector Regressor)

from sklearn.svm import SVR

svr_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_model.fit(x_train, y_train)
```

```python
y_pred_svr = svr_model.predict(x_test)
y_pred_svr
```

```python
mae_svr = mean_absolute_error(y_test, y_pred_svr)
mse_svr = mean_squared_error(y_test, y_pred_svr)
r2_svr = r2_score(y_test, y_pred_svr)
```

```python
print(f"Mean Absolute Error (MAE): {mae_svr}")
print(f"Root Mean Squared Error (RMSE): {mse_svr}")
print(f"R-squared (R2): {r2_svr}")
```

```python
# 5. Ridge Regression

from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x_train, y_train)
```

```python
y_pred_ridge = ridge_model.predict(x_test)
y_pred_ridge
```

```python
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
```

```python
print(f"Mean Absolute Error (MAE): {mae_ridge}")
print(f"Root Mean Squared Error (RMSE): {mse_ridge}")
print(f"R-squared (R2): {r2_ridge}")
```

```python
# 6. Lasso Regression

from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=1.0)
lasso_model.fit(x_train, y_train)
```

```python
y_pred_lasso = lasso_model.predict(x_test)
y_pred_lasso
```

```python
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
```

```python
print(f"Mean Absolute Error (MAE): {mae_lasso}")
print(f"Root Mean Squared Error (RMSE): {mse_lasso}")
print(f"R-squared (R2): {r2_lasso}")
```

```python
# 6. Neural Network (Multi-Layer Perceptron - MLP)

from sklearn.neural_network import MLPRegressor

mlp_model = MLPRegressor(hidden_layer_sizes=(100,100), max_iter=500, random_state=42)
mlp_model.fit(x_train, y_train)
```

```python
y_pred_mlp = mlp_model.predict(x_test)
y_pred_mlp
```

```python
mae_mlp = mean_absolute_error(y_test, y_pred_mlp)
mse_mlp = mean_squared_error(y_test, y_pred_mlp)
r2_mlp = r2_score(y_test, y_pred_mlp)
```

```python
print(f"Mean Absolute Error (MAE): {mae_mlp}")
print(f"Root Mean Squared Error (RMSE): {mse_mlp}")
print(f"R-squared (R2): {r2_mlp}")
```

```python
# linear regression
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression(n_jobs=-1)
lr_model.fit(x_train, y_train)
```

```python
y_pred_lr = lr_model.predict(x_test)
y_pred_lr
```

```python
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)
```

```python
print(f"Mean Absolute Error (MAE): {mae_lr}")
print(f"Root Mean Squared Error (RMSE): {mse_lr}")
print(f"R-squared (R2): {r2_lr}")
```

```python
# COMPARISION OF MAE, MSE AND R2 SCORE OF ALL THE ABOVE MODELS TO UNDERSTAND THE BEST-FIT MODEL

print(f"Random Forest -        MAE: {mae_rf},   RMSE: {mse_rf}, R2: {r2_rf}")
print(f"XGBoost -              MAE: {mae_xgb},  RMSE: {mse_xgb}, R2: {r2_xgb}")
print(f"LightGBM -             MAE: {mae_lgbm}, RMSE: {mse_lgbm},  R2: {r2_lgbm}")
print(f"SVR -                  MAE: {mae_svr},  RMSE: {mse_svr}, R2: {r2_svr}")
print(f"Ridge Regression -     MAE: {mae_ridge},   RMSE: {mse_ridge},  R2: {r2_ridge}")
print(f"Lasso Regression -     MAE: {mae_lasso},     RMSE: {mse_lasso},   R2: {r2_lasso}")
print(f"Neural Network (MLP) - MAE: {mae_mlp},  RMSE: {mse_mlp}, R2: {r2_mlp}")
print(f"Linear Regression - MAE: {mae_lr}, RMSE: {mse_lr}, R2: {r2_lr}")
```
